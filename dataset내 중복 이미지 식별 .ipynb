```python
def find_duplication(img1, img2):
    image = cv2.imread(img1)
    image = cv2.resize(image, (224, 224))

    template = cv2.imread(img2)
    template = cv2.resize(template, (224, 224))
    result = cv2.matchTemplate(image, template, cv2.TM_CCOEFF_NORMED)
    threshold = 0.95

    return result
```

```python
folders = ['lounge sofa', 'bar chair', 'module sofa', 'round stool chair', 'recliner sofa', 'bean bag sofa', 'folding chair', 'chesterfield sofa', 'dining chair', 'gaming chair', 'office_chair']
```

```python
image_lst = []
template_lst = []

for f in os.listdir(os.getcwd() + '/dataset/train/' + folders[0]):
    image_lst.append(os.getcwd() + '/dataset/train/' + folders[0] + '/' + f)
for f in os.listdir(os.getcwd() + '/dataset/test/' + folders[0]):
    template_lst.append(os.getcwd() + '/dataset/test/' + folders[0] + '/' + f)
```

```python
for tem in template_lst:
    count = 0
    for img in image_lst:
        result = find_duplication(img, tem)
        if result[0][0] > 0.95:
            print(img, tem, '는 중복된 파일입니다.')
            count += 1

    if count == 0:
        print(tem, '와 중복된 파일이 없습니다.')
```

```python
for folder in folders[2:]:
    count = 0
    image_lst = []
    template_lst = []

    for f in os.listdir(os.getcwd() + '/dataset/train/' + folder):
        image_lst.append(os.getcwd() + '/dataset/train/' + folder + '/' + f)
    for f in os.listdir(os.getcwd() + '/dataset/test/' + folder):
        template_lst.append(os.getcwd() + '/dataset/test/' + folder + '/' + f)

    for tem in template_lst:
        for img in image_lst:
            cnt = 0
            try:
                result = find_duplication(img, tem)
                if result[0][0] > 0.95:
                    print(img, tem, '는 중복된 파일입니다.')
                    count += 1
            except:
                pass

    if count == 0:
        print(folder, '폴더에는 중복된 파일이 없습니다.')
```
