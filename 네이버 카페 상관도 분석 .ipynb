```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Okt
import numpy as np
```

```python
def okt_returns(df, column):
  okt_nouns = df[column].apply(lambda x: okt.nouns(x))
  return okt_nouns
```

```python
def okt_tokenizer(text):
    return okt.morphs(text)
```

```python
def tfidf_vect_maker(df, column):
  tfidf_vect = TfidfVectorizer(tokenizer = okt_tokenizer, ngram_range = (1, 2), min_df = 3, max_df = 0.9)
  tfidf_vect.fit(df[column])
  tfidf_matrix = tfidf_vect.transform(df[column])

  return tfidf_vect, tfidf_matrix
```

```python
def pickle_save(tfidf_vect, tfidf_matrix):
  pickle.dump(tfidf_vect, open('tfidf.pickle', 'wb'))
  pickle.dump(tfidf_matrix, open('tfidf_matrix.pickle', 'wb'))
```

```python
def find_sim_title(df, sorted_ind, title_name, top_n = 11):
  titles = df[df['title'] == title_name]
  title_index = titles.index.values
  similar_indexes = sorted_ind[title_index, :(top_n)]
  return df.iloc[similar_indexes[0]]
```
