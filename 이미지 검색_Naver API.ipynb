```python
def image_crawling(keyword):
    encText = urllib.parse.quote(keyword)

    result_list = []

    for i in range(1, 1000, 100):
        url = "https://openapi.naver.com/v1/search/image?query=" + encText + "&display=100&sort=sim&start=" + str(i)  # JSON 결과

        request = urllib.request.Request(url)
        request.add_header("X-Naver-Client-Id",client_id)
        request.add_header("X-Naver-Client-Secret",client_secret)

        response = urllib.request.urlopen(request)
        rescode = response.getcode()

        if(rescode==200):
            response_body = response.read()
            response_json = json.loads(response_body)

            result_data = response_json['items']

            for temp in result_data:
                image_data = {}

                image_title = temp['title']
                image_thumbnail = temp['thumbnail']
                image_link = temp['link']

                image_data['title'] = image_title
                image_data['thumbnail'] = image_thumbnail
                image_data['link'] = image_link

                result_list.append(image_data)

        else:
            print("Error Code:" + rescode)

    image = pd.DataFrame.from_dict(result_list)

    return image
```

```python
def save_file_to_csv(df):
    file_name = input('확장자명을 제외한 파일 이름을 입력하세요: ')
    df.to_csv(os.path.join(os.getcwd(), file_name + '.csv'), index = False)
    time.sleep(10)
```

```python
def save_images(df, query):
    # 폴더가 없다면 query에 입력한 이름의 폴더를 생성
    if not os.path.isdir('./' + query + '/'):
        os.makedirs('./' + query + '/')

    for i, url in enumerate(df['link']):
        path = os.getcwd() + '/' + query + '/image_' + str(i) + '.jpg'
        req = urllib.request.Request(url, headers={'User-Agent': 'Mozilla/5.0'})
        response = urllib.request.urlopen(req)
        with open(path, 'wb') as file:
            file.write(response.read())
        time.sleep(4)
```
