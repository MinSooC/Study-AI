```python
# 문자열의 인덱싱 : 맨 앞의 데이터는 0번째 데이터
min = '870124-1065478'

year = min[0] + min[1]
birth = min[2] + min[3] + min[4] + min[5]
gender = min[7]

print(year)
print(birth)
print(gender)
```

```python
# 문자열의 슬라이싱
# [a:b]는 a 이상 b 미만의 인덱스를 슬라이싱하라는 뜻
year = min[:2]
birth = min[2:6]

print(year)
print(birth)
```

```python
# 맨 마지막, 즉 뒤에서 첫번째로 있는 데이터를 가져오기
data1 = 'a12364F'
data2 = 'b17284686M'

data1[-1]
```

```python
# 주민등록번호의 암호화  ex) '870124-*******'
min = '870124-1065478'

min_new = min[:7] + '*' * 7
min_new
```

```python
# 문자열 Formatting : 문자나 숫자를 중간중간에 대입하고 싶을 때
# 숫자를 대입하기 (%d : % 뒤에 나오는 정수형의 숫자 데이터를 %d의 자리에 집어넣어라)
'Height : %d' % 165
```

```python
# (%f : % 뒤에 나오는 실수형의 숫자 데이터를 %f 자리에 집어넣어라)
'num : %f' % 7.24
```

```python
# 문자를 대입하기 (%s : % 뒤에 나오는 문자 데이터를 %s의 자리에 집어넣어라)
'Name : %s' % '홍길동'
```

```python
# 여러개의 변수를 대입
text = '홍길동'
num = 27
'Name : %s, Age : %d' % (text, num)
```

```python
# 문자열 Format 함수
# 해당 방식의 특징 : 중괄호 안의 데이터의 타입이 상관없다.
'나이 : {0}'.format(30)
```

```python
'나이 : {0}'.format('30')
```

```python
# 중괄호 안의 숫자에 따라 순차적으로 데이터를 대입
'나이 : {0}, {1}'.format(30, '25')
```

```python
# 변수 형태로도 사용 가능 : 이 때, 순서에 상관없이 데이터를 대입할 수 있다.
'나이 : {a}, {b}'.format(b = '25', a = 30)
```

```python
# 문자열 함수
a = 0.123456

# 소수점 둘째자리까지 표현하기 (5의 경우, 길이를 5글자로 제한하는 것, 만약 부족할 시 공백을 통해서라도 채움)
'{0:5.2f}'.format(a)
```

#**딕셔너리 자료형**

```python
# 딕셔너리는 Key와 Value로 구성
# 중괄호({})를 사용하며, {} 안에 콤마(,)를 이용하여 데이터를 생성
# Key와 Value는 콜론(:)으로 구분하며, Key는 중복되지 않게 생성

ai_data = {'name':['홍길동', '임꺽정'], 'age':[32, 43]}
ai_data['name']  # 'name'이라는 데이터의 키값에 해당하는 value를 출력
```

```python
# 딕셔너리 데이터 내부의 Key 값 확인
ai_data.keys()
```

```python
# 딕셔너리 데이터 내부의 Value 값 확인
ai_data.values()
```

```python
# 비어있는 딕셔너리에 데이터 추가
data = {}

data['name'] = ['홍길동', '임꺽정']
data['age'] = [32, 43]
data
```

#**조건문**

```python
# 조건문이란, 자료형에서 다룬 데이터를 어떠한 조건에 맞춰 판단한 후 처리할 때 사용
# 문제를 풀기 위해 생각해야 할 것 :
#  1. 변수를 어떻게 선언할 것인가?
#  2. 조건문을 어떻게 만들 것인가?

# 비교연산자와 조건연산자 : 조건문에서 조건의 참이나 거짓이 되는 조건을 판단하기 위해 사용
```

```python
score = 100

# 버그 발생 시 라인이 실행되었는지 여부를 하나씩 확인할 것
if score == 100:
    print('축하합니다.')
```

```python
# 아이디와 비밀번호가 일치하면 로그인이 되었습니다.를  출력하세요.
id = 'ai'
pw = 'ai'
list = [id, pw]

if list == ['ai', 'ai']:
    print('로그인이 되었습니다.')
```

```python
id = 'ai'
pw = 'ai'
count = 0

if id == 'ai':
    count += 1

if pw == 'ai':
    count += 1

if count == 2:
    print('로그인이 되었습니다.')
```
